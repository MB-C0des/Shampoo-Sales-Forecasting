```python
# Define a function to reformat the Month column

def reformat_date(month):
    year_suffix = month[:2]
    month_abbr = month[3:]
    year = 2020 + int(year_suffix)
    return f"{year}-{month_abbr}"


df['Month'] = df['Month'].apply(reformat_date)


print(df)
```

```python
# Prepare data

df['Sales'] = df['Sales'].astype(float)

X = df['Sales'].values  # Extract the series values
size = int(len(X) * 0.66)  # Define the split size
train, test = X[:size], X[size:]  # Split into training and test sets
history = list(train)  # Initialize history with the training data
predictions = []  # Initialize predictions list
differences = []  # Initialize differences list

# Rolling forecast
for t in range(len(test)):
    # Fit ARIMA model on the historical data
    model = ARIMA(history, order=(5,1,0))  # Example ARIMA parameters (p,d,q)
    model_fit = model.fit()
    
    # Forecast the next value
    output = model_fit.forecast(steps=1)
    yhat = output[0]  # Extract the forecasted value
    
    # Append the prediction and actual observation
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)  # Update history with the observed value
    print('predicted=%f, expected=%f' % (yhat, obs))

    
   
```

```python

# Plot the rolling forecast ARIMA model
plt.plot(test)
plt.plot(predictions, color='red')
plt.legend(['Expected','Predicted'])
plt.show()
```

```python
# Print all values predicts/ expected

 
print("\nDetailed Results:")
print("Index\tPredicted\tExpected")
for i in range(len(test)):
    print(f'{i}\t{predictions[i]:.3f}\t\t{test[i]:.3f}')
```

```python
# Forecast future values
future_steps = 4  # Number of future periods to forecast
future_forecast = model_fit.forecast(steps=future_steps)
print(f'\nForecasted values for the next {future_steps} periods: {future_forecast}')
```

```python
df_train = df[df['ds'] < '2023-11']
df_test = df[df['ds'] >= '2023-11']

df_train
```

```python
model_prophet  = Prophet()
model_prophet.fit(df_train)
```

```python
df_future = model_prophet.make_future_dataframe(periods=12, freq='MS')
display(df_future)



```

```python
forecast_prophet = model_prophet.predict(df_future)
forecast_prophet[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].round()
```

```python
# plot the time series 
forecast_plot = model_prophet.plot(forecast_prophet)

# add a vertical line at the end of the training period
axes = forecast_plot.gca()
last_training_date = forecast_prophet['ds'].iloc[-12]
axes.axvline(x=last_training_date, color='red', linestyle='--', label='Training End')

# plot true test data for the period after the red line
df_test['ds'] = pd.to_datetime(df_test['ds'])
plt.plot(df_test['ds'], df_test['y'],'ro', markersize=3, label='True Test Data')

# show the legend to distinguish between the lines
plt.legend()
```
